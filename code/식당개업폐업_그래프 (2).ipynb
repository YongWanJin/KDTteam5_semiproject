```python
def project_array(coord, p1_type, p2_type):
    """
    좌표계 변환 함수
    - coord: x, y 좌표 정보가 담긴 NumPy Array
    - p1_type: 입력 좌표계 정보 ex) epsg:5179
    - p2_type: 출력 좌표계 정보 ex) epsg:4326
    """
    p1 = pyproj.Proj(init=p1_type)
    p2 = pyproj.Proj(init=p2_type)
    fx, fy = pyproj.transform(p1, p2, coord[:, 0], coord[:, 1])
    return np.dstack([fx, fy])[0]
```

```python
### 구별 전분기 대비 폐업률 집계

index = [gu for gu in df_all_res['시군구명'].unique()]
columns = ["2019년4분기", "2020년1분기", "2020년2분기", "2020년3분기", "2020년4분기",
           "2021년1분기", "2021년2분기", "2021년3분기", "2021년4분기", "2022년1분기"]
inc_gu = pd.DataFrame(index = index, columns = columns)  

# 증가율 집계
for p in range( len(df_all_res['시점_구분_변수'].unique())-1 ) :

    means_before = df_all_res[df_all_res['시점_구분_변수']==p ].groupby('시군구명')['폐업_률'].mean()
    means_before = np.array(means_before) 

    means_now = df_all_res[df_all_res['시점_구분_변수']==p+1 ].groupby('시군구명')['폐업_률'].mean()
    means_now = np.array(means_now) 

    inc = (means_now - means_before)/means_before*100 
    inc = np.round(inc, 2).reshape(25,1)
    inc = inc.tolist()
    inc_gu.iloc[:,[p]] = inc # 입력

display(inc_gu) # 결과
```

```python
### 구별 전분기 대비 개업율 집계

index = [gu for gu in df_all_res['시군구명'].unique()]
columns = ["2019년4분기", "2020년1분기", "2020년2분기", "2020년3분기", "2020년4분기",
           "2021년1분기", "2021년2분기", "2021년3분기", "2021년4분기", "2022년1분기"]
open_gu = pd.DataFrame(index = index, columns = columns)  

# 증가율 집계
for p in range( len(df_all_res['시점_구분_변수'].unique())-1 ) :

    means_before = df_all_res[df_all_res['시점_구분_변수']==p ].groupby('시군구명')['개업_율'].mean()
    means_before = np.array(means_before) 

    means_now = df_all_res[df_all_res['시점_구분_변수']==p+1 ].groupby('시군구명')['개업_율'].mean()
    means_now = np.array(means_now) 

    open = (means_now - means_before)/means_before*100 
    open = np.round(open, 2).reshape(25,1)
    open = open.tolist()
    open_gu.iloc[:,[p]] = open # 입력

display(open_gu) # 결과
```
