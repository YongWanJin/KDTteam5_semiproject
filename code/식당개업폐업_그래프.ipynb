```python
# 19년도 1,2 분기 데이터 Drop

df_19 = pd.read_csv('/content/drive/MyDrive/서울시_우리마을가게_상권분석서비스(신_상권배후지_점포)_2019년.csv',header=0, encoding='CP949')

df_19_1 = df_19['기준_분기_코드'] < 3
#df_19[df_19_1]
df_19.drop(index = df_19[df_19_1].index, inplace=True)
```

```python
# 22년도 파일에서 20, 21년 데이터 Drop

df_22 = pd.read_csv('/content/drive/MyDrive/서울시 우리마을가게 상권분석서비스(상권배후지-점포)_2022년.csv',header=0, encoding='CP949')

df_22_1 = df_22['기준_년_코드'] < 2022
#df_22[df_22_1]
df_22.drop(index = df_22[df_22_1].index, inplace = True)
```

```python
df_20 = pd.read_csv('/content/drive/MyDrive/서울시_우리마을가게_상권분석서비스(신_상권배후지_점포)_2020년.csv',header=0, encoding='CP949')
df_21 = pd.read_csv('/content/drive/MyDrive/서울시_우리마을가게_상권분석서비스(신_상권배후지_점포)_2021년.csv',header=0, encoding='CP949')
```

```python
df_all = pd.concat([df_19, df_20, df_21, df_22])
```

```python
df_all.to_csv("서울시_우리마을가게_상권분석서비스(신_상권배후지_점포)_19-22년.csv")
```
